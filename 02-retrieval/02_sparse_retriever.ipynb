# Sparse Retriever
> BM25-based keyword retrieval for exact term matching.

| Module | `anchor.retrieval` |
|--------|--------------------|
| Key classes | `SparseRetriever`, `InMemoryContextStore`, `QueryBundle` |
| Difficulty | Beginner |

`SparseRetriever` uses the BM25 algorithm for term-frequency-based retrieval.
It excels at exact keyword matching where dense embeddings may miss specific terms.

**Time:** ~5 minutes

## Setup

```python
from anchor.retrieval import SparseRetriever
from anchor.storage import InMemoryContextStore
from anchor.models import ContextItem, SourceType, QueryBundle
```

## Create the Sparse Retriever
Only a context store is needed -- no embedding function required.

```python
ctx_store = InMemoryContextStore()
retriever = SparseRetriever(context_store=ctx_store)

print(f"Retriever type: {type(retriever).__name__}")
```

## Index Documents

```python
items = [
    ContextItem(id="bm25-1", content="Python is great for data science and machine learning.",
               source=SourceType.RETRIEVAL, score=0.0, priority=5, token_count=12),
    ContextItem(id="bm25-2", content="Rust provides memory safety without a garbage collector.",
               source=SourceType.RETRIEVAL, score=0.0, priority=5, token_count=10),
    ContextItem(id="bm25-3", content="Go was designed at Google for systems programming.",
               source=SourceType.RETRIEVAL, score=0.0, priority=5, token_count=10),
    ContextItem(id="bm25-4", content="JavaScript powers the interactive web and Node.js servers.",
               source=SourceType.RETRIEVAL, score=0.0, priority=5, token_count=11),
    ContextItem(id="bm25-5", content="Machine learning models need large datasets for training.",
               source=SourceType.RETRIEVAL, score=0.0, priority=5, token_count=10),
]

retriever.index(items)
print(f"Indexed {len(items)} documents for BM25")
```

## Retrieve with Keyword Query
BM25 scores documents based on term frequency and inverse document frequency.

```python
query = QueryBundle(query_str="machine learning")
results = retriever.retrieve(query, top_k=3)

print(f"Query: '{query.query_str}'")
print(f"Results ({len(results)}):\n")
for i, item in enumerate(results):
    print(f"  [{i+1}] score={item.score:.4f}  {item.content[:60]}")
```

## Compare Queries
Different keywords surface different documents.

```python
queries = ["memory safety", "data science", "Google systems", "web servers"]

for q in queries:
    results = retriever.retrieve(QueryBundle(query_str=q), top_k=2)
    top = results[0].content[:50] if results else "(none)"
    print(f"  '{q}' -> {top}")
```

## Key Takeaways
- `SparseRetriever` implements BM25 keyword-based retrieval
- No embedding function needed -- operates on raw token frequencies
- Best for queries with specific, distinctive terms
- Complements `DenseRetriever` -- combine them in a `HybridRetriever`
