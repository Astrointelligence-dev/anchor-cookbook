# Recipe 03 — Table Extraction
> Parse and encode tabular data from Markdown and HTML sources.

| | |
|---|---|
| **Module** | `anchor.multimodal.tables` / `anchor.multimodal.encoders` |
| **Key classes** | `MarkdownTableParser`, `HTMLTableParser`, `TableEncoder` |
| **Difficulty** | Beginner |

```python
# Setup
from anchor.multimodal.tables import MarkdownTableParser, HTMLTableParser
from anchor.multimodal.encoders import TableEncoder
from anchor.multimodal import MultiModalContent, ModalityType
```

## Walkthrough

```python
# 1 — Parse a Markdown table
md_parser = MarkdownTableParser()

md_table = (
    "| Name  | Age | Role      |\n"
    "|-------|-----|-----------|\n"
    "| Alice | 30  | Engineer  |\n"
    "| Bob   | 25  | Designer  |\n"
    "| Carol | 35  | Manager   |"
)

parsed = md_parser.parse(md_table)

print(f"Headers : {parsed.headers}")
print(f"Rows    : {len(parsed.rows)}")
for row in parsed.rows:
    print(f"  {row}")
```

```python
# 2 — Parse an HTML table
html_parser = HTMLTableParser()

html_table = (
    "<table>"
    "<tr><th>Name</th><th>Age</th><th>Role</th></tr>"
    "<tr><td>Alice</td><td>30</td><td>Engineer</td></tr>"
    "<tr><td>Bob</td><td>25</td><td>Designer</td></tr>"
    "</table>"
)

parsed_html = html_parser.parse(html_table)

print(f"Headers : {parsed_html.headers}")
print(f"Rows    : {len(parsed_html.rows)}")
for row in parsed_html.rows:
    print(f"  {row}")
```

```python
# Compare the two parsers
print(f"Markdown headers: {parsed.headers}")
print(f"HTML headers    : {parsed_html.headers}")
print(f"Markdown rows   : {len(parsed.rows)}")
print(f"HTML rows       : {len(parsed_html.rows)}")
```

```python
# 3 — Encode tables with TableEncoder
# TableEncoder converts table content into a text representation
table_encoder = TableEncoder(format="markdown")
print(f"Encoder : {type(table_encoder).__name__}")
print(f"Format  : markdown")
```

```python
# Wrap the markdown table in MultiModalContent
table_content = MultiModalContent(
    modality=ModalityType.TABLE,
    content=md_table,
)

encoded = table_encoder.encode(table_content)
print(f"Encoded table:\n{encoded}")
```

```python
# Verify encoding produces a non-empty string
print(f"Encoded type   : {type(encoded).__name__}")
print(f"Encoded length : {len(encoded)} chars")
print(f"Non-empty      : {len(encoded) > 0}")
```

```python
# 4 — Larger table example
# A more complex table with numeric data
sales_table = (
    "| Quarter | Revenue  | Growth |\n"
    "|---------|----------|--------|\n"
    "| Q1 2025 | $1.2M    | 15%    |\n"
    "| Q2 2025 | $1.5M    | 25%    |\n"
    "| Q3 2025 | $1.8M    | 20%    |\n"
    "| Q4 2025 | $2.1M    | 17%    |"
)

sales_parsed = md_parser.parse(sales_table)
print(f"Headers: {sales_parsed.headers}")
print(f"Rows   : {len(sales_parsed.rows)}")
for row in sales_parsed.rows:
    print(f"  {row}")
```

```python
# Encode the sales table
sales_content = MultiModalContent(
    modality=ModalityType.TABLE,
    content=sales_table,
)
sales_encoded = table_encoder.encode(sales_content)
print(f"Encoded sales table:\n{sales_encoded}")
```

```python
# Verify both parsers handle edge cases consistently
simple_md = "| A | B |\n|---|---|\n| 1 | 2 |"
simple_html = "<table><tr><th>A</th><th>B</th></tr><tr><td>1</td><td>2</td></tr></table>"

md_result = md_parser.parse(simple_md)
html_result = html_parser.parse(simple_html)

print(f"Markdown parse: headers={md_result.headers}, rows={md_result.rows}")
print(f"HTML parse    : headers={html_result.headers}, rows={html_result.rows}")
```

## Key Takeaways
- `MarkdownTableParser` and `HTMLTableParser` extract structured headers and rows.
- `TableEncoder` converts table content into encoded text for downstream use.
- Both parsers produce a consistent parsed representation with `.headers` and `.rows`.
- Tables are treated as a first-class `ModalityType.TABLE` in the multimodal system.
