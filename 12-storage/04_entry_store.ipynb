# Entry Store
> Manage `MemoryEntry` objects for long-term memory.

`InMemoryEntryStore` stores `MemoryEntry` instances with CRUD operations.
Memory entries represent facts, preferences, or other persistent knowledge
that should survive across sessions.

**Time:** ~5 minutes

## Setup

```python
from anchor.storage import InMemoryEntryStore
from anchor.models import MemoryEntry
```

## Create an Entry Store

```python
store = InMemoryEntryStore()

print(f"Type: {type(store).__name__}")
```

## Add Memory Entries
Each entry carries an ID, content, relevance score, and optional metadata.

```python
entry1 = MemoryEntry(
    id="mem-1",
    content="User prefers dark mode interfaces.",
    relevance_score=0.8,
    metadata={"category": "preference"},
)

entry2 = MemoryEntry(
    id="mem-2",
    content="User is a senior Python developer.",
    relevance_score=0.9,
    metadata={"category": "profile"},
)

entry3 = MemoryEntry(
    id="mem-3",
    content="Last project used FastAPI and PostgreSQL.",
    relevance_score=0.6,
    metadata={"category": "history"},
)

store.add(entry1)
store.add(entry2)
store.add(entry3)

print("Added 3 memory entries")
```

## List All Entries
`.list_all()` returns every stored entry.

```python
all_entries = store.list_all()

print(f"Total entries: {len(all_entries)}\n")
for e in all_entries:
    print(f"  [{e.id}] score={e.relevance_score:.1f}  "
          f"cat={e.metadata.get('category', 'N/A')}")
    print(f"          '{e.content}'")
```

## Update an Entry
Modify an entry and call `.update()` to persist the change.

```python
# Update the relevance score
entry1.relevance_score = 0.95
entry1.content = "User strongly prefers dark mode interfaces."

store.update(entry1)
print("Updated 'mem-1'")

# Verify
updated = [e for e in store.list_all() if e.id == "mem-1"][0]
print(f"  New score:   {updated.relevance_score}")
print(f"  New content: '{updated.content}'")
```

## Delete an Entry

```python
store.delete("mem-3")
print("Deleted 'mem-3'")

remaining = store.list_all()
print(f"Remaining: {len(remaining)} entries")
for e in remaining:
    print(f"  [{e.id}] '{e.content[:50]}'")
```

## Key Takeaways
- `InMemoryEntryStore` manages `MemoryEntry` objects
- `.add()`, `.update()`, `.list_all()`, `.delete()` cover full CRUD
- Entries carry relevance scores for prioritized retrieval
- Use metadata to categorize entries (preference, profile, history, etc.)
