# Document Store
> Store, retrieve, and manage text documents with metadata.

`InMemoryDocumentStore` provides a simple key-value store for text documents.
Each document has an ID, content string, and optional metadata dictionary.

**Time:** ~5 minutes

## Setup

```python
from anchor.storage import InMemoryDocumentStore
```

## Create a Document Store

```python
store = InMemoryDocumentStore()

print(f"Type: {type(store).__name__}")
```

## Add Documents
Each document requires an ID, content, and optional metadata.

```python
store.add_document(
    doc_id="doc-1",
    content="Python is a high-level programming language.",
    metadata={"author": "Alice", "category": "programming"},
)

store.add_document(
    doc_id="doc-2",
    content="Rust focuses on safety and performance.",
    metadata={"author": "Bob", "category": "programming"},
)

store.add_document(
    doc_id="doc-3",
    content="Machine learning is a subset of artificial intelligence.",
    metadata={"author": "Carol", "category": "AI"},
)

print("Added 3 documents")
```

## Retrieve a Single Document
`.get_document()` returns the content string by ID (or `None` if missing).

```python
content = store.get_document("doc-1")

print(f"ID:      doc-1")
print(f"Content: {content}")
```

## List All Document IDs
`.list_documents()` returns a list of stored document IDs.

```python
doc_ids = store.list_documents()

print(f"Total documents: {len(doc_ids)}\n")
for doc_id in doc_ids:
    content = store.get_document(doc_id)
    print(f"  [{doc_id}] {content[:50]}...")
```

## Delete a Document

```python
store.delete_document("doc-1")
print("Deleted 'doc-1'")

remaining = store.list_documents()
print(f"Remaining: {len(remaining)} documents")
for doc_id in remaining:
    content = store.get_document(doc_id)
    print(f"  [{doc_id}] {content[:50]}")
```

## Key Takeaways
- `InMemoryDocumentStore` is a simple ID-keyed text store
- `.add_document()` stores content with metadata; `.get_document()` retrieves by ID
- `.list_documents()` lists all IDs; `.delete_document()` removes one
- Ideal for prototyping before moving to a persistent backend
