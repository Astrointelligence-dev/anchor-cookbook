# Context Store
> Manage `ContextItem` objects with scores, priorities, and source types.

`InMemoryContextStore` holds `ContextItem` instances — the core unit of
context in Anchor.  Each item carries content, a source type, relevance
score, priority, and token count.

**Time:** ~5 minutes

## Setup

```python
from anchor.storage import InMemoryContextStore
from anchor.models import ContextItem, SourceType
```

## Create a Context Store

```python
store = InMemoryContextStore()

print(f"Type: {type(store).__name__}")
```

## Add Context Items
Build `ContextItem` objects and store them.

```python
item1 = ContextItem(
    id="ctx-1",
    content="Python was created by Guido van Rossum.",
    source=SourceType.RETRIEVAL,
    score=0.95,
    priority=5,
    token_count=8,
)

item2 = ContextItem(
    id="ctx-2",
    content="The user prefers concise answers.",
    source=SourceType.MEMORY,
    score=0.80,
    priority=7,
    token_count=6,
)

item3 = ContextItem(
    id="ctx-3",
    content="Always respond in English.",
    source=SourceType.SYSTEM,
    score=1.0,
    priority=10,
    token_count=4,
)

store.add(item1)
store.add(item2)
store.add(item3)

print("Added 3 context items")
```

## Retrieve a Single Item

```python
retrieved = store.get("ctx-1")

print(f"ID:       {retrieved.id}")
print(f"Content:  {retrieved.content}")
print(f"Source:   {retrieved.source.value}")
print(f"Score:    {retrieved.score}")
print(f"Priority: {retrieved.priority}")
print(f"Tokens:   {retrieved.token_count}")
```

## List All Items

```python
all_items = store.get_all()

print(f"Total items: {len(all_items)}\n")
for item in all_items:
    print(f"  [{item.id}] source={item.source.value:10s}  "
          f"score={item.score:.2f}  priority={item.priority}  "
          f"tokens={item.token_count}")
    print(f"           content='{item.content}'")
```

## Sort by Priority and Score
Context items are plain objects — sort them for pipeline ordering.

```python
sorted_items = sorted(all_items, key=lambda x: (-x.priority, -x.score))

print("Items sorted by priority (desc), then score (desc):\n")
for item in sorted_items:
    print(f"  [{item.id}] priority={item.priority}  "
          f"score={item.score:.2f}  '{item.content[:40]}'")
```

## Key Takeaways
- `InMemoryContextStore` holds `ContextItem` objects
- Each item has content, source type, score, priority, and token count
- `.add()`, `.get()`, and `.get_all()` cover basic CRUD
- Items can be sorted by priority/score for pipeline ordering
