# JSON File Store
> Persist memory entries to disk as JSON.

`JsonFileMemoryStore` extends the entry store concept with file-based
persistence.  Entries are saved to and loaded from a JSON file, making it
easy to maintain memory across application restarts.

**Time:** ~5 minutes

## Setup

```python
import os
import tempfile
from anchor.storage import JsonFileMemoryStore
from anchor.models import MemoryEntry
```

## Create a Store with a File Path
We use a temporary file so this notebook is self-contained.

```python
tmp_dir = tempfile.mkdtemp()
file_path = os.path.join(tmp_dir, "test_memory.json")

store = JsonFileMemoryStore(file_path=file_path)

print(f"File path: {file_path}")
print(f"File exists: {os.path.exists(file_path)}")
```

## Add Entries and Save to Disk

```python
entry1 = MemoryEntry(
    id="mem-1",
    content="User prefers concise code examples.",
    relevance_score=0.85,
    metadata={"source": "conversation"},
)

entry2 = MemoryEntry(
    id="mem-2",
    content="Primary language is Python 3.12.",
    relevance_score=0.9,
    metadata={"source": "profile"},
)

store.add(entry1)
store.add(entry2)

# Persist to file
store.save()

print(f"Saved {len(store.list_all())} entries")
print(f"File exists: {os.path.exists(file_path)}")
print(f"File size:   {os.path.getsize(file_path)} bytes")
```

## Load from Disk
Create a new store instance pointing to the same file and load data.

```python
loaded = JsonFileMemoryStore(file_path=file_path)
loaded.load()

entries = loaded.list_all()
print(f"Loaded {len(entries)} entries from disk\n")
for e in entries:
    print(f"  [{e.id}] score={e.relevance_score}  '{e.content}'")
```

## Modify and Re-Save
Updates are persisted by calling `.save()` again.

```python
new_entry = MemoryEntry(
    id="mem-3",
    content="Frequently works with async/await patterns.",
    relevance_score=0.7,
    metadata={"source": "observation"},
)

loaded.add(new_entry)
loaded.save()

print(f"Now {len(loaded.list_all())} entries on disk")
print(f"File size: {os.path.getsize(file_path)} bytes")
```

## Cleanup
Remove the temporary file.

```python
os.remove(file_path)
os.rmdir(tmp_dir)
print(f"Cleaned up temporary file: {file_path}")
```

## Key Takeaways
- `JsonFileMemoryStore` persists memory entries as JSON on disk
- `.save()` writes current state; `.load()` restores from file
- Combine with `MemoryEntry` for simple, file-based persistence
- Ideal for local development and small-scale deployments
