# Human Evaluation
> Collect and manage human judgments alongside automated metrics.

`HumanEvaluationCollector` stores qualitative feedback from human
reviewers, letting you combine expert judgment with algorithmic
scores for a complete evaluation picture.

**Time:** ~5 minutes

## Setup

```python
from anchor.evaluation import HumanEvaluationCollector
from anchor.evaluation.models import HumanJudgment
```

## Create a Collector
The collector acts as an in-memory store for human judgments.

```python
collector = HumanEvaluationCollector()

print(f"Collector type: {type(collector).__name__}")
print(f"Judgments so far: {len(collector.get_judgments())}")
```

## Add a Single Judgment
`HumanJudgment` captures who evaluated, what they saw, their
numeric rating, and free-text feedback.

```python
judgment = HumanJudgment(
    evaluator_id="reviewer-1",
    query="What is prompt engineering?",
    answer="Prompt engineering is the practice of designing inputs to LLMs.",
    rating=4,
    feedback="Good answer, covers the core concept clearly.",
)

collector.add_judgment(judgment)

print(f"Added judgment from: {judgment.evaluator_id}")
print(f"Rating:   {judgment.rating}")
print(f"Feedback: {judgment.feedback}")
```

## Add Multiple Judgments
Simulate several reviewers scoring different query-answer pairs.

```python
more_judgments = [
    HumanJudgment(
        evaluator_id="reviewer-1",
        query="What is RAG?",
        answer="RAG combines retrieval with generation.",
        rating=5,
        feedback="Excellent -- concise and accurate.",
    ),
    HumanJudgment(
        evaluator_id="reviewer-2",
        query="What is prompt engineering?",
        answer="Prompt engineering is the practice of designing inputs to LLMs.",
        rating=3,
        feedback="Correct but could be more detailed.",
    ),
    HumanJudgment(
        evaluator_id="reviewer-2",
        query="What are embeddings?",
        answer="Embeddings map words to numbers.",
        rating=2,
        feedback="Too vague, missing key details.",
    ),
    HumanJudgment(
        evaluator_id="reviewer-3",
        query="What is RAG?",
        answer="RAG combines retrieval with generation.",
        rating=4,
        feedback="Solid answer, could mention accuracy benefits.",
    ),
]

for j in more_judgments:
    collector.add_judgment(j)

print(f"Total judgments: {len(collector.get_judgments())}")
```

## Retrieve All Judgments
`get_judgments()` returns the full list.

```python
judgments = collector.get_judgments()

print(f"{'Reviewer':<12} {'Query':<30} {'Rating':>6}")
print("-" * 50)
for j in judgments:
    print(f"{j.evaluator_id:<12} {j.query:<30} {j.rating:>6}")
```

## Compute Summary Statistics
Basic aggregation across all collected judgments.

```python
ratings = [j.rating for j in judgments]
avg_rating = sum(ratings) / len(ratings)

reviewers = set(j.evaluator_id for j in judgments)
queries = set(j.query for j in judgments)

print(f"Total judgments:  {len(judgments)}")
print(f"Unique reviewers: {len(reviewers)}")
print(f"Unique queries:   {len(queries)}")
print(f"Average rating:   {avg_rating:.2f}")
print(f"Min rating:       {min(ratings)}")
print(f"Max rating:       {max(ratings)}")
```

## Per-Reviewer Breakdown
See how each reviewer scores on average.

```python
from collections import defaultdict

by_reviewer = defaultdict(list)
for j in judgments:
    by_reviewer[j.evaluator_id].append(j.rating)

print(f"{'Reviewer':<12} {'Count':>6} {'Avg':>6}")
print("-" * 26)
for reviewer, scores in sorted(by_reviewer.items()):
    avg = sum(scores) / len(scores)
    print(f"{reviewer:<12} {len(scores):>6} {avg:>6.2f}")
```

## Key Takeaways

- `HumanEvaluationCollector` provides a simple API for gathering
  qualitative ratings alongside automated metrics.
- `HumanJudgment` captures evaluator ID, query, answer, numeric
  rating, and free-text feedback.
- Combine human evaluation with retrieval and RAG metrics for a
  complete quality picture.
- Analyse inter-reviewer agreement to gauge evaluation reliability.
