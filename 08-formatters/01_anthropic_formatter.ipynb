# Anthropic Formatter
> Format a ContextWindow into the Anthropic Messages API structure.

`AnthropicFormatter` converts your assembled context into the dict
format expected by the Anthropic Python SDK (`system` + `messages`).

**Time:** ~5 minutes

## Setup

```python
from anchor.formatters import AnthropicFormatter
from anchor.models import ContextWindow, ContextItem, SourceType
```

## Build a Sample Context Window
Create a window with a system item and a retrieval item so we can
see how each source type maps to the Anthropic format.

```python
from anchor.models import ContextWindow, ContextItem, SourceType

# Build a sample context window with mixed sources
window = ContextWindow(max_tokens=4096)
window.add_item(ContextItem(
    id="sys-1",
    content="You are helpful.",
    source=SourceType.SYSTEM,
    score=1.0,
    priority=10,
    token_count=5,
))
window.add_item(ContextItem(
    id="doc-1",
    content="Python is a language.",
    source=SourceType.RETRIEVAL,
    score=0.9,
    priority=5,
    token_count=6,
))

print(f"Context window: {len(window.items)} items, "
      f"budget {window.max_tokens} tokens")
for item in window.items:
    print(f"  [{item.source.value}] {item.id}: {item.content!r}")
```

## Format for the Anthropic API
The formatter exposes a `format_type` property and a `format()` method
that returns the provider-specific payload.

```python
formatter = AnthropicFormatter()
print(f"Format type: {formatter.format_type}")
```

```python
output = formatter.format(window)
print(type(output))
print(output)
```

## Inspect the Output Structure
The Anthropic format separates system-level content from user/assistant
messages, matching the Messages API contract.

```python
if isinstance(output, dict):
    for key, value in output.items():
        print(f"{key}: {value}")
else:
    # String output — print directly
    print(output)
```

## Key Takeaways

- `AnthropicFormatter` produces output ready for the Anthropic Messages API.
- System-source items are routed to the `system` field.
- All other items appear in the `messages` list.
- Use `format_type` to identify the formatter at runtime.
