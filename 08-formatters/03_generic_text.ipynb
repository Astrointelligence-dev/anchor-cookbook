# Generic Text Formatter
> Format a ContextWindow as a plain-text string.

`GenericTextFormatter` renders context as a human-readable text
block. Useful for logging, debugging, or providers that accept
raw text prompts.

**Time:** ~5 minutes

## Setup

```python
from anchor.formatters import GenericTextFormatter
from anchor.models import ContextWindow, ContextItem, SourceType
```

## Build a Sample Context Window

```python
from anchor.models import ContextWindow, ContextItem, SourceType

# Build a sample context window with mixed sources
window = ContextWindow(max_tokens=4096)
window.add_item(ContextItem(
    id="sys-1",
    content="You are helpful.",
    source=SourceType.SYSTEM,
    score=1.0,
    priority=10,
    token_count=5,
))
window.add_item(ContextItem(
    id="doc-1",
    content="Python is a language.",
    source=SourceType.RETRIEVAL,
    score=0.9,
    priority=5,
    token_count=6,
))

print(f"Context window: {len(window.items)} items, "
      f"budget {window.max_tokens} tokens")
for item in window.items:
    print(f"  [{item.source.value}] {item.id}: {item.content!r}")
```

## Format as Plain Text
The output is a single string — no dicts, no nesting.

```python
formatter = GenericTextFormatter()
print(f"Format type: {formatter.format_type}")
```

```python
output = formatter.format(window)
print(type(output))
print()
print(output)
```

## Use Cases
Plain text is handy for:
- **Debugging** — quickly inspect what the model will see.
- **Logging** — write context snapshots to log files.
- **Simple providers** — any API that takes a single prompt string.

```python
# Example: write context to a log file
import tempfile, os

log_path = os.path.join(tempfile.gettempdir(), "context_debug.txt")
with open(log_path, "w") as f:
    f.write(output)

print(f"Wrote {len(output)} chars to {log_path}")
print(f"First 120 chars: {output[:120]!r}...")
```

## Key Takeaways

- `GenericTextFormatter` returns a plain `str`.
- Ideal for debugging, logging, or text-only LLM providers.
- Same `Formatter` protocol as the provider-specific formatters.
