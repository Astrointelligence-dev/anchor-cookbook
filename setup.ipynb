## Quick Smoke Test
Build a minimal context pipeline to verify everything works.

```python
from anchor.pipeline import ContextPipeline
from anchor.models import ContextItem, SourceType, QueryBundle
from anchor.formatters import GenericTextFormatter

# Create a minimal pipeline
pipeline = ContextPipeline(max_tokens=4096)
pipeline.add_system_prompt("You are a helpful assistant.")
pipeline.with_formatter(GenericTextFormatter())

# Build context
query = QueryBundle(query_str="Hello, world!")
result = pipeline.build(query)

print(f"Items in window: {len(result.window.items)}")
print(f"Token utilization: {result.window.utilization:.1%}")
print(f"\nFormatted output preview:")
print(result.formatted_output[:200] if isinstance(result.formatted_output, str) else str(result.formatted_output)[:200])
print("\n--- Setup complete! Your environment is ready. ---")
```

## Key Takeaways
- `astro-anchor[all]` installs all optional dependencies
- `ContextPipeline` is the main entry point for building context
- Use `GenericTextFormatter` for plain text output
