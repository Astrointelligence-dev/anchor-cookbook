# Graph Memory
> Track entities and relationships across conversations.

`SimpleGraphMemory` provides a lightweight in-memory entity graph. You can
store entities with metadata, create typed relationships between them, and
link entities to specific memory entry IDs for retrieval.

**Time:** ~5 minutes

## Setup

```python
from anchor.memory import SimpleGraphMemory
```

## Create Entities
Entities are nodes in the graph. Each has an ID and optional metadata.

```python
graph = SimpleGraphMemory()

graph.add_entity("alice", metadata={"role": "engineer", "team": "backend"})
graph.add_entity("bob", metadata={"role": "designer", "team": "frontend"})
graph.add_entity("project-x", metadata={"type": "project", "status": "active"})
graph.add_entity("project-y", metadata={"type": "project", "status": "planned"})

print("Entities added: alice, bob, project-x, project-y")
```

## Create Relationships
Relationships are directed edges with a type label.

```python
graph.add_relationship("alice", "works_on", "project-x")
graph.add_relationship("bob", "works_on", "project-x")
graph.add_relationship("alice", "mentors", "bob")
graph.add_relationship("alice", "leads", "project-y")

print("Relationships:")
print("  alice --works_on--> project-x")
print("  bob   --works_on--> project-x")
print("  alice --mentors---> bob")
print("  alice --leads-----> project-y")
```

## Link Memory Entries to Entities
Associate memory entry IDs with entities so you can retrieve relevant memories.

```python
graph.link_memory("alice", "mem-001")
graph.link_memory("alice", "mem-002")
graph.link_memory("project-x", "mem-003")

print("Linked memories:")
print("  alice     -> mem-001, mem-002")
print("  project-x -> mem-003")
```

## Traverse the Graph
`get_related_entities` performs a breadth-first traversal up to `max_depth`.

```python
related = graph.get_related_entities("alice", max_depth=2)

print(f"Entities related to 'alice' (depth <= 2): {len(related)}")
for entity_id in related:
    print(f"  - {entity_id}")
```

## Retrieve Related Memory IDs
Collect all memory IDs linked to entities within the traversal radius.

```python
memory_ids = graph.get_related_memory_ids("alice", max_depth=2)

print(f"Memory IDs related to 'alice': {memory_ids}")
```

## Key Takeaways
- `SimpleGraphMemory` stores entities, relationships, and memory links
- Use `add_relationship()` for directed, typed edges between entities
- `link_memory()` associates entries with entities for retrieval
- `get_related_entities()` traverses the graph up to a specified depth
- Graph memory shines when conversations involve multiple interacting concepts
