# Memory Manager
> Unified facade for conversation memory and persistent storage.

`MemoryManager` combines short-term conversation memory with a persistent
entry store. It handles adding messages, building context items, and recalling
relevant entries from long-term storage.

**Time:** ~5 minutes

## Setup

```python
from anchor.memory import MemoryManager, SlidingWindowMemory
from anchor.storage import InMemoryEntryStore
```

## Initialize the Manager
Provide a token budget and a persistent store backend.

```python
store = InMemoryEntryStore()

manager = MemoryManager(
    conversation_tokens=4096,
    persistent_store=store,
)

print(f"Conversation token budget: 4096")
print(f"Persistent store type: {type(store).__name__}")
```

## Add Conversation Messages

```python
manager.add_message("user", "Tell me about Python")
manager.add_message("assistant", "Python is a high-level programming language "
    "known for its readability and versatility.")
manager.add_message("user", "What about its package manager?")
manager.add_message("assistant", "pip is the standard package manager for "
    "Python. It installs packages from PyPI.")

print("Added 4 messages to conversation memory.")
```

## Get Context Items
Combine conversation turns and any relevant persistent entries into context items.

```python
context_items = manager.get_context_items()

print(f"Total context items: {len(context_items)}")
for item in context_items:
    print(f"  [{item.source_type.value}] {str(item.content)[:70]}")
```

## Recall from Persistent Storage
Search the persistent store for entries matching a query.

```python
recalled = manager.recall("Python")

print(f"Recalled entries: {len(recalled) if recalled else 0}")
if recalled:
    for entry in recalled:
        print(f"  - {str(entry)[:80]}")
else:
    print("  (No persistent entries stored yet -- "
          "recall returns results from long-term storage.)")
```

## Key Takeaways
- `MemoryManager` is the recommended high-level interface for memory
- It combines conversation (short-term) and persistent (long-term) memory
- `.get_context_items()` produces pipeline-ready context items
- `.recall(query)` searches the persistent store for relevant memories
- Swap `InMemoryEntryStore` for a vector DB store in production
