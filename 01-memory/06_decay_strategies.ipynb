# Decay Strategies
> Model how memory fades over time using decay curves and recency scoring.

Anchor provides two decay models for persistent memory entries and two
recency scorers for conversation turns.

**Time:** ~5 minutes

## Setup

```python
from anchor.memory.decay import EbbinghausDecay, LinearDecay
from anchor.memory.recency import ExponentialRecencyScorer, LinearRecencyScorer
```

## Ebbinghaus Forgetting Curve
Models the classic exponential forgetting curve. Repeated access (reinforcement)
slows the decay.

```python
ebbinghaus = EbbinghausDecay(
    base_strength=1.0,
    reinforcement_factor=0.5,
)

# Simulate retention at various hours since last access
print(f"{'Hours':>6}  {'Retention':>10}")
print("-" * 20)
for hours in [0, 1, 6, 24, 72, 168]:
    # Approximate by computing decay manually for demonstration
    import math
    retention = math.exp(-hours / (24 * ebbinghaus.base_strength))
    print(f"{hours:>6}  {retention:>10.4f}")
```

## Linear Decay
Retention decreases linearly, reaching 0.5 at `half_life_hours`.

```python
linear = LinearDecay(half_life_hours=168.0)  # 1 week

print(f"Half-life: {linear.half_life_hours} hours (1 week)")
print(f"\n{'Hours':>6}  {'Retention':>10}")
print("-" * 20)
for hours in [0, 24, 72, 168, 336, 504]:
    retention = max(0.0, 1.0 - (hours / (2 * linear.half_life_hours)))
    print(f"{hours:>6}  {retention:>10.4f}")
```

## Exponential Recency Scorer
Scores conversation turns by position. Recent turns score higher with
exponential drop-off.

```python
exp_scorer = ExponentialRecencyScorer(decay_rate=2.0)

total_turns = 10
print(f"Exponential scorer (decay_rate={exp_scorer.decay_rate})")
print(f"\n{'Position':>8}  {'Score':>8}")
print("-" * 20)
for pos in range(total_turns):
    import math
    # position 0 = most recent
    score = math.exp(-exp_scorer.decay_rate * pos / total_turns)
    print(f"{pos:>8}  {score:>8.4f}")
```

## Linear Recency Scorer
Scores decrease linearly from 1.0 (most recent) to near 0.0 (oldest).

```python
lin_scorer = LinearRecencyScorer()

total_turns = 10
print("Linear recency scorer")
print(f"\n{'Position':>8}  {'Score':>8}")
print("-" * 20)
for pos in range(total_turns):
    score = 1.0 - (pos / total_turns)
    print(f"{pos:>8}  {score:>8.4f}")
```

## Key Takeaways
- **EbbinghausDecay**: exponential forgetting with reinforcement support
- **LinearDecay**: simpler linear drop-off controlled by `half_life_hours`
- **ExponentialRecencyScorer** / **LinearRecencyScorer**: score conversation
  turns by position for priority-aware context building
- Decay strategies feed into garbage collection (next notebook)
