# Basic Agent Configuration
> Set up an Anchor Agent with a model, system prompt, and memory.

The `Agent` class is the high-level entry point for agentic AI
applications. It wraps the context pipeline, Anthropic SDK, and
tool-use loop into a single fluent interface.

**Time:** ~5 minutes

## Setup

```python
from anchor.agent import Agent
from anchor.memory import SlidingWindowMemory
```

## Create an Agent
The `Agent` constructor accepts a model identifier and optional
parameters for token limits, retry behaviour, and round caps.

```python
agent = Agent(
    model="claude-haiku-4-5-20251001",
    max_response_tokens=2048,
    max_rounds=10,
)

print(f"Agent created")
print(f"  model:              claude-haiku-4-5-20251001")
print(f"  max_response_tokens: 2048")
print(f"  max_rounds:          10")
```

## Set a System Prompt
`.with_system_prompt()` returns `self` for fluent chaining.

```python
agent.with_system_prompt("You are a helpful assistant.")

print("System prompt set:", bool(agent._system_prompt))
print("Prompt text:", repr(agent._system_prompt))
```

## Attach Memory
`SlidingWindowMemory` keeps recent conversation turns within a
token budget. The agent uses it to maintain conversation context.

```python
memory = SlidingWindowMemory(max_tokens=4096)

print(f"Memory type:   {type(memory).__name__}")
print(f"Token budget:  {memory.max_tokens}")
print(f"Current turns: {len(memory.turns)}")
```

## Wire Memory into the Agent
`.with_memory()` connects the memory to the underlying pipeline
so conversation history is included in every API call.

```python
agent.with_memory(memory)

print("Memory attached:", agent.memory is not None)
print("Pipeline has memory:", agent.pipeline is not None)
```

## Inspect the Pipeline
The agent exposes its underlying `ContextPipeline` for
debugging and introspection.

```python
pipeline = agent.pipeline

print(f"Pipeline type: {type(pipeline).__name__}")
print(f"Has formatter:  True")  # AnthropicFormatter is set by default
print(f"Last result:   {agent.last_result}")
```

## Key Takeaways
- `Agent` combines pipeline + API client + tool loop in one class
- `.with_system_prompt()`, `.with_memory()`, `.with_tools()` return `self` for chaining
- `SlidingWindowMemory` is the simplest memory backend for agents
- `agent.chat(msg)` is an iterator that yields streaming text chunks
- Access the pipeline via `agent.pipeline` and last build via `agent.last_result`
