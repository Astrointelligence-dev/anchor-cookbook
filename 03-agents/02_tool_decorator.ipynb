# Tool Decorator
> Turn plain Python functions into agent-callable tools with a single decorator.

The `@tool` decorator auto-generates an `AgentTool` from the function's
type hints and docstring. Tools can be exported to Anthropic, OpenAI, or
generic schema formats.

**Time:** ~5 minutes

## Setup

```python
from anchor.agent import tool, AgentTool
```

## Basic `@tool` Decorator
Apply `@tool` to auto-generate the tool name, description, and
JSON Schema from the function signature.

```python
@tool
def calculate_sum(a: int, b: int) -> int:
    """Add two numbers together."""
    return a + b

print(f"Type: {type(calculate_sum).__name__}")
print(f"Name: {calculate_sum.name}")
print(f"Description: {calculate_sum.description}")
```

## `@tool` with Explicit Name and Description
Override the auto-generated name and description when needed.

```python
@tool(name="greet", description="Generate a greeting")
def greet_user(name: str) -> str:
    """Greet a user by name."""
    return f"Hello, {name}!"

print(f"Name: {greet_user.name}")        # 'greet' (overridden)
print(f"Description: {greet_user.description}")  # 'Generate a greeting'
```

## Inspect the Generated Schema
The `input_schema` property contains the JSON Schema derived from
type hints.

```python
import json

print("=== calculate_sum input_schema ===")
print(json.dumps(calculate_sum.input_schema, indent=2))
```

## Export to Provider Formats
`AgentTool` can export to Anthropic, OpenAI, or a generic format.

```python
print("=== Anthropic Schema ===")
print(json.dumps(calculate_sum.to_anthropic_schema(), indent=2))

print("\n=== OpenAI Schema ===")
print(json.dumps(calculate_sum.to_openai_schema(), indent=2))
```

## Direct Invocation
The original function is accessible via the `.fn` attribute.

```python
# Call the underlying function directly
result = calculate_sum.fn(a=3, b=5)
print(f"calculate_sum(3, 5) = {result}")

greeting = greet_user.fn(name="Alice")
print(f"greet('Alice') = {greeting}")
```

## Input Validation
`AgentTool.validate_input()` checks tool input against the schema.

```python
# Valid input
ok, err = calculate_sum.validate_input({"a": 3, "b": 5})
print(f"Valid input:   ok={ok}, err={err!r}")

# Missing required field
ok, err = calculate_sum.validate_input({"a": 3})
print(f"Missing field: ok={ok}, err={err!r}")

# Wrong type
ok, err = calculate_sum.validate_input({"a": "three", "b": 5})
print(f"Wrong type:    ok={ok}, err={err!r}")
```

## Key Takeaways
- `@tool` auto-generates `AgentTool` from function signature + docstring
- Override `name` and `description` with `@tool(name=..., description=...)`
- `.input_schema` holds the JSON Schema for the tool's parameters
- `.to_anthropic_schema()` / `.to_openai_schema()` export to provider formats
- `.fn` gives access to the raw callable; `.validate_input()` checks schema compliance
