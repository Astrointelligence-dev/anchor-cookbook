# Skills System
> Organise tools into named, discoverable skill groups with on-demand activation.

Skills let you bundle related tools together and control when they
become available to the agent. *Always* skills are loaded immediately;
*on_demand* skills are advertised but only activated when the agent
explicitly requests them.

**Time:** ~7 minutes

## Setup

```python
from anchor.agent import Skill, SkillRegistry, tool, AgentTool
```

## Define Tools for the Skill
Create a pair of tools that will be grouped into a single skill.

```python
@tool
def search_docs(query: str) -> str:
    """Search documentation."""
    return f"Results for: {query}"

@tool
def save_note(content: str) -> str:
    """Save a note."""
    return f"Saved: {content}"

print(f"Tools created: {search_docs.name}, {save_note.name}")
```

## Create a Skill
A `Skill` bundles tools with a name, description, instructions,
activation mode, and optional tags.

```python
research_skill = Skill(
    name="research",
    description="Research and note-taking tools",
    instructions="Use these tools to research topics and save notes.",
    tools=(search_docs, save_note),
    activation="on_demand",
    tags=("research", "notes"),
)

print(f"Skill name:        {research_skill.name}")
print(f"Description:       {research_skill.description}")
print(f"Activation:        {research_skill.activation}")
print(f"Tags:              {research_skill.tags}")
print(f"Tools in skill:    {len(research_skill.tools)}")
for t in research_skill.tools:
    print(f"  - {t.name}: {t.description}")
```

## Skill Registry
`SkillRegistry` manages registration, activation, and tool discovery.

```python
registry = SkillRegistry()
registry.register(research_skill)

print(f"Registered skills: {list(registry._skills.keys())}")
print(f"Is 'research' active? {registry.is_active('research')}")
print(f"On-demand skills: {[s.name for s in registry.on_demand_skills()]}")
```

## Activate an On-Demand Skill
On-demand skills become active only after explicit activation.

```python
activated = registry.activate("research")

print(f"Activated: {activated.name}")
print(f"Is 'research' active now? {registry.is_active('research')}")
```

## Retrieve Active Tools
`active_tools()` returns all tools from currently-active skills.

```python
active_tools = registry.active_tools()

print(f"Active tools: {len(active_tools)}")
for t in active_tools:
    print(f"  - {t.name}: {t.description}")
```

## Skill Discovery Prompt
The registry generates a prompt listing available on-demand skills
for the agent's system message.

```python
# Deactivate to see the discovery prompt
registry.deactivate("research")
discovery = registry.skill_discovery_prompt()

print("Discovery prompt sent to the agent:")
print(discovery)
```

## Always-On Skills
Skills with `activation="always"` are active from the moment
they are registered.

```python
@tool
def get_time() -> str:
    """Get the current time."""
    return "2026-03-13T10:30:00Z"

utility_skill = Skill(
    name="utility",
    description="General utility tools",
    tools=(get_time,),
    activation="always",
)

registry2 = SkillRegistry()
registry2.register(utility_skill)

print(f"Is 'utility' active? {registry2.is_active('utility')}")
print(f"Active tools: {[t.name for t in registry2.active_tools()]}")
```

## Built-in Skills
Anchor ships pre-built memory and RAG skills.

```python
from anchor.agent import memory_skill, rag_skill, memory_tools, rag_tools

print("Built-in skills available:")
print(f"  memory_skill: {type(memory_skill).__name__}")
print(f"  rag_skill:    {type(rag_skill).__name__}")
print(f"  memory_tools: callable that returns list[AgentTool]")
print(f"  rag_tools:    callable that returns list[AgentTool]")
```

## Key Takeaways
- `Skill` bundles related tools with metadata and activation mode
- `SkillRegistry` manages registration, activation, and discovery
- `activation="on_demand"` skills are advertised but loaded lazily
- `activation="always"` skills are active from registration
- `active_tools()` returns tools from all currently-active skills
- Anchor provides built-in `memory_skill` and `rag_skill`
