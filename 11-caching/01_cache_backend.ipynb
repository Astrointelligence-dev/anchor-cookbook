# Cache Backend Protocol
> Understand the caching interface and implement a custom backend.

`CacheBackend` is a protocol that defines four methods: `get()`, `set()`,
`delete()`, and `clear()`.  Any object implementing these methods can serve
as a cache in Anchor — no inheritance required.

**Time:** ~5 minutes

## Setup

```python
from anchor.protocols import CacheBackend
```

## The CacheBackend Protocol
The protocol defines a minimal key-value cache interface:
- `get(key)` – retrieve a cached value (or `None`)
- `set(key, value, ttl=None)` – store a value with optional TTL
- `delete(key)` – remove a single entry
- `clear()` – wipe the entire cache

```python
# Inspect the protocol methods
print("CacheBackend protocol methods:")
for method in ["get", "set", "delete", "clear"]:
    print(f"  - {method}()")
```

## Build a Custom Redis-Like Cache
A stub implementation showing the required method signatures.

```python
class RedisCache:
    """Stub that mimics a Redis-backed cache."""

    def __init__(self):
        self._store = {}

    def get(self, key: str):
        value = self._store.get(key)
        print(f"GET '{key}' -> {value}")
        return value

    def set(self, key: str, value, ttl=None):
        self._store[key] = value
        ttl_msg = f" (ttl={ttl}s)" if ttl else ""
        print(f"SET '{key}' = {value}{ttl_msg}")

    def delete(self, key: str) -> bool:
        removed = self._store.pop(key, None) is not None
        print(f"DEL '{key}' -> {'removed' if removed else 'not found'}")
        return removed

    def clear(self):
        count = len(self._store)
        self._store.clear()
        print(f"CLEAR -> removed {count} entries")


cache = RedisCache()
print(f"Type: {type(cache).__name__}")
```

## Verify Protocol Conformance
`isinstance()` confirms the class satisfies the protocol.

```python
is_valid = isinstance(cache, CacheBackend)
print(f"Satisfies CacheBackend protocol: {is_valid}")

assert is_valid, "RedisCache must implement CacheBackend"
```

## Exercise the Custom Cache
Run through a typical set/get/delete/clear cycle.

```python
# Store values
cache.set("user:1", {"name": "Alice", "role": "admin"})
cache.set("user:2", {"name": "Bob", "role": "viewer"}, ttl=300)

# Retrieve
print()
cache.get("user:1")
cache.get("user:99")  # miss

# Delete one entry
print()
cache.delete("user:2")
cache.delete("user:99")  # no-op

# Clear everything
print()
cache.clear()
```

## Key Takeaways
- `CacheBackend` is a four-method protocol: `get`, `set`, `delete`, `clear`
- Any class matching the signature is accepted — no base class needed
- Use `isinstance()` for runtime protocol verification
- Custom backends can wrap Redis, Memcached, DynamoDB, or any store
