# Budget Presets
> Ready-made token budgets for common application patterns.

Anchor ships three factory functions that create sensible default budgets
for chat, RAG, and agent workloads.  Use them as starting points and
customize allocations as needed.

**Time:** ~3 minutes

## Setup

```python
from anchor.models import (
    default_chat_budget,
    default_rag_budget,
    default_agent_budget,
)
```

## Chat Budget (8K tokens)
Optimized for conversational assistants with moderate context.

```python
chat = default_chat_budget()

print(f"Chat budget: {chat.total_tokens} tokens")
print(f"Reserve:     {chat.reserve_tokens} tokens")
print(f"Allocations:")
for alloc in chat.allocations:
    print(f"  {alloc.source.value:12s}  max={alloc.max_tokens:>5}  "
          f"priority={alloc.priority}  overflow={alloc.overflow_strategy}")
```

## RAG Budget (12K tokens)
Larger retrieval allocation for knowledge-grounded answers.

```python
rag = default_rag_budget()

print(f"RAG budget: {rag.total_tokens} tokens")
print(f"Reserve:    {rag.reserve_tokens} tokens")
print(f"Allocations:")
for alloc in rag.allocations:
    print(f"  {alloc.source.value:12s}  max={alloc.max_tokens:>5}  "
          f"priority={alloc.priority}  overflow={alloc.overflow_strategy}")
```

## Agent Budget (16K tokens)
Extra room for tool results, planning traces, and multi-step reasoning.

```python
agent = default_agent_budget()

print(f"Agent budget: {agent.total_tokens} tokens")
print(f"Reserve:      {agent.reserve_tokens} tokens")
print(f"Allocations:")
for alloc in agent.allocations:
    print(f"  {alloc.source.value:12s}  max={alloc.max_tokens:>5}  "
          f"priority={alloc.priority}  overflow={alloc.overflow_strategy}")
```

## Compare All Three
Quick side-by-side summary.

```python
for name, b in [("chat", chat), ("rag", rag), ("agent", agent)]:
    sources = ", ".join(a.source.value for a in b.allocations)
    print(f"{name:6s}  total={b.total_tokens:>6}  "
          f"reserve={b.reserve_tokens:>4}  sources=[{sources}]")
```

## Key Takeaways
- `default_chat_budget()` → 8K tokens, balanced for conversation
- `default_rag_budget()` → 12K tokens, heavy retrieval allocation
- `default_agent_budget()` → 16K tokens, room for tool results
- All presets return a regular `TokenBudget` you can modify after creation
