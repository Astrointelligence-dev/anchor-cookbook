# Custom Tokenizer
> Implement the `Tokenizer` protocol with your own counting logic.

Anchor defines a `Tokenizer` protocol with two methods: `count_tokens()`
and `truncate_to_tokens()`.  Any class that implements these methods can be
used anywhere Anchor expects a tokenizer — no inheritance required.

**Time:** ~5 minutes

## Setup

```python
from anchor.protocols import Tokenizer
```

## The Tokenizer Protocol
The protocol requires exactly two methods:
- `count_tokens(text: str) -> int`
- `truncate_to_tokens(text: str, max_tokens: int) -> str`

```python
# Inspect the protocol
print("Tokenizer protocol methods:")
for name in ["count_tokens", "truncate_to_tokens"]:
    print(f"  - {name}")
```

## Build a Simple Word-Based Tokenizer
A lightweight tokenizer that counts whitespace-separated words.

```python
class SimpleTokenizer:
    """Counts tokens as whitespace-separated words."""

    def count_tokens(self, text: str) -> int:
        return len(text.split())

    def truncate_to_tokens(self, text: str, max_tokens: int) -> str:
        words = text.split()
        return " ".join(words[:max_tokens])


tokenizer = SimpleTokenizer()
print(f"Type: {type(tokenizer).__name__}")
```

## Verify Protocol Conformance
`isinstance()` checks work with `runtime_checkable` protocols.

```python
is_valid = isinstance(tokenizer, Tokenizer)
print(f"Satisfies Tokenizer protocol: {is_valid}")

assert is_valid, "SimpleTokenizer must satisfy the Tokenizer protocol"
```

## Use the Custom Tokenizer
Exercise both methods with sample text.

```python
text = "The quick brown fox jumps over the lazy dog"

count = tokenizer.count_tokens(text)
print(f"Text:       '{text}'")
print(f"Word count: {count}")

truncated = tokenizer.truncate_to_tokens(text, max_tokens=5)
print(f"\nTruncated to 5 tokens: '{truncated}'")
print(f"Truncated count:       {tokenizer.count_tokens(truncated)}")
```

## Build a Character-Based Tokenizer
Another custom implementation — this time counting individual characters.

```python
class CharTokenizer:
    """Counts tokens as individual characters."""

    def count_tokens(self, text: str) -> int:
        return len(text)

    def truncate_to_tokens(self, text: str, max_tokens: int) -> str:
        return text[:max_tokens]


char_tok = CharTokenizer()

print(f"Satisfies Tokenizer: {isinstance(char_tok, Tokenizer)}")
print(f"Char count of 'Hello': {char_tok.count_tokens('Hello')}")
print(f"Truncated to 3: '{char_tok.truncate_to_tokens('Hello', 3)}'")
```

## Key Takeaways
- The `Tokenizer` protocol is structural — implement two methods, no base class
- `isinstance()` verification works at runtime
- Custom tokenizers plug into budgets, counters, and pipelines seamlessly
- Word-based, character-based, or any counting scheme can be used
