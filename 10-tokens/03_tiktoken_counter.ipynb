# Tiktoken Counter
> Accurate token counting and truncation using OpenAI’s tiktoken.

`TiktokenCounter` wraps the tiktoken library to provide fast, accurate
token counting compatible with GPT-3.5/4 models.  Anchor also exposes a
convenience function `get_default_counter()` for the most common encoding.

**Time:** ~5 minutes

## Setup

```python
from anchor.tokens import TiktokenCounter, get_default_counter
```

## Create a Counter
Specify the tiktoken encoding name (e.g., `cl100k_base` for GPT-4).

```python
counter = TiktokenCounter(encoding_name="cl100k_base")

print(f"Encoding: cl100k_base")
print(f"Type:     {type(counter).__name__}")
```

## Count Tokens
Pass any string to `.count_tokens()` to get the exact token count.

```python
text = "Hello, world!"
count = counter.count_tokens(text)

print(f"Text:   '{text}'")
print(f"Tokens: {count}")

# Try a longer string
long_text = "The quick brown fox jumps over the lazy dog. " * 5
print(f"\nLong text tokens: {counter.count_tokens(long_text)}")
```

## Truncate to Token Limit
`.truncate_to_tokens()` cuts text so it fits within a maximum token count.

```python
original = "This is a longer sentence that we want to truncate to a small number of tokens."

truncated = counter.truncate_to_tokens(original, max_tokens=5)

print(f"Original ({counter.count_tokens(original)} tokens):")
print(f"  {original}")
print(f"\nTruncated ({counter.count_tokens(truncated)} tokens):")
print(f"  {truncated}")
```

## Default Counter
`get_default_counter()` returns a pre-configured counter suitable for most
use cases.

```python
default = get_default_counter()

print(f"Default counter type: {type(default).__name__}")
print(f"Token count test:     {default.count_tokens('Anchor is great!')}")
```

## Counting Different Content Types
Token counts vary by content — code, prose, and special characters all
tokenize differently.

```python
samples = [
    "Hello, world!",
    "def fibonacci(n): return n if n <= 1 else fibonacci(n-1) + fibonacci(n-2)",
    "こんにちは世界",
    "https://docs.anchor.dev/api/tokens",
]

for s in samples:
    tokens = counter.count_tokens(s)
    print(f"  {tokens:>3} tokens │ {s[:50]}")
```

## Key Takeaways
- `TiktokenCounter` gives exact token counts matching OpenAI models
- `.truncate_to_tokens()` safely trims content to a budget
- `get_default_counter()` is the quickest way to get started
- Token counts vary by language, formatting, and content type
